```python
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv("./Cost_of_Living_Index_by_Country_2024.csv")
sns.set_style("whitegrid")
sns.set(rc={'figure.figsize': (11, 7)})
```

```python
# ==============================================================
# 1. Best Value Countries  (Horizontal bar plot)
# ==============================================================
df['Value_Score'] = df['Local Purchasing Power Index'] / df['Cost of Living Plus Rent Index']
best_value = df.nlargest(12, 'Value_Score').sort_values('Value_Score', ascending=True)  # highest at bottom

ax3 = sns.barplot(data=best_value, y='Country', x='Value_Score', palette='Blues_d')

ax3.set_title('3. Best Value Countries: Most Lifestyle per Dollar\n(Longer bar = better value – highest score is #1)', fontsize=16, fontweight='bold')
ax3.set_xlabel('Value Score = Purchasing Power ÷ Total Cost')
ax3.set_ylabel('')

# Place value labels INSIDE the bars near the end (clear and close to the visual)
for container in ax3.containers:
    ax3.bar_label(container, fmt='%.2f', label_type='edge', padding=5, fontweight='bold', fontsize=12)
```

```python
# ==============================================================
# 2. Distribution of total cost (Histogram with KDE)
# ==============================================================
ax1 = sns.histplot(data=df, x='Cost of Living Plus Rent Index', bins=18, color='#ff9999', kde=True, edgecolor='black')
ax1.set_title('1. Most of the World Is Much Cheaper Than New York', fontsize=16, fontweight='bold')
ax1.set_xlabel('Cost of Living + Rent Index (NYC = 100)')
ax1.set_ylabel('Number of Countries')
ax1.axvline(100, color='red', linestyle='--', linewidth=2, label='New York City = 100')
ax1.legend()
```

```python
#==============================================================
# 3. Countries with highest rent burden (Cleveland dot plot)
#==============================================================

df['Rent_Burden'] = df['Rent Index'] / df['Cost of Living Index']
top_rent_burden = df.nlargest(12, 'Rent_Burden').sort_values('Rent_Burden')

ax2 = sns.stripplot(data=top_rent_burden, y='Country', x='Rent_Burden', 
                    color='#d35400', size=14, jitter=False, marker='o')
ax2.set_title('2. Countries Where Rent Hurts the Most\n(Rent Index ÷ Daily Costs)', fontsize=16, fontweight='bold')
ax2.set_xlabel('Rent Burden Ratio (higher = rent takes bigger share)')
ax2.set_ylabel('')

for i, v in enumerate(top_rent_burden['Rent_Burden']):
    ax2.text(v + 0.01, i, f'{v:.2f}', va='center', fontweight='bold', color='black')
```

```python
#==============================================================
# 4. Spread of costs across ALL countries (beeswarm plot)
#============================================================== 
spread_data = df.melt(value_vars=[
    'Groceries Index', 
    'Restaurant Price Index', 
    'Cost of Living Index', 
    'Rent Index', 
    'Cost of Living Plus Rent Index', 
    'Local Purchasing Power Index'
], var_name='Cost Type', value_name='Index Value')

spread_data['Cost Type'] = spread_data['Cost Type'].str.replace(' Index', '')

ax3 = sns.stripplot(data=spread_data, x='Index Value', y='Cost Type', 
                    color='#9b59b6', size=8, jitter=True, alpha=0.7)

ax3.axvline(100, color='red', linewidth=3, linestyle='--', label='New York City = 100')

ax3.set_title('3. How Spread Out Costs Are Across All Countries\n(Dots = countries, wider spread = more variety)', fontsize=18, fontweight='bold')
ax3.set_xlabel('Index Value (NYC = 100)')
ax3.set_ylabel('')
ax3.set_xlim(0, 200)
ax3.legend()
```

```python
#========================================================================
# 5.salaries-groceries-rent (scatter plot with color mapping)
#========================================================================
ax4 = sns.scatterplot(data=df, 
                      x='Groceries Index', 
                      y='Local Purchasing Power Index', 
                      hue='Rent Index', 
                      palette='viridis',   # smooth, clear color gradient
                      s=100,               # good size, not too big/small
                      alpha=0.85, 
                      edgecolor='white', 
                      linewidth=1)

ax4.set_title('4. Salary Strength vs Groceries Cost\n(Darker color = more expensive rent)', fontsize=18, fontweight='bold')
ax4.set_xlabel('Groceries Index (lower = cheaper food, NYC = 100)')
ax4.set_ylabel('Local Purchasing Power Index (higher = stronger salaries)')

# NYC reference lines
ax4.axvline(100, color='gray', linestyle='--', alpha=0.8)
ax4.axhline(100, color='gray', linestyle='--', alpha=0.8)

ax4.legend(title='Rent Index', loc='upper left', bbox_to_anchor=(1.02, 1), fontsize=12)
```
